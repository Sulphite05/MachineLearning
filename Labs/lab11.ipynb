## Q2.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
from sklearn import datasets

iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features_standardized = scaler.fit_transform(features)
```

```python
from sklearn.decomposition import KernelPCA

# Create support vector classifier
pca = KernelPCA(n_components=2, kernel="rbf", gamma=15)
# Train model
model = pca.fit_transform(features_standardized)
```

```python
print("Original Features:", features_standardized.shape[1])
print("Reduced Features:", model.shape[1])
```

## Q3.

```python
from sklearn.model_selection import train_test_split

trainX1, testX1, trainY1, testY1 = train_test_split(features, target, random_state=2, test_size=0.2)
```

## Support Vector Machine(Without PCA)

```python
svc1 = SVC(kernel="rbf", C=100, gamma=10)
svc1.fit(trainX1, trainY1)
```

```python
predY_ = svc1.predict(testX1)
```

```python
print("The accuracy is "+str(metrics.accuracy_score(testY1, predY_)*100)+"%")
cmatrix_=confusion_matrix(testY1, predY_)
print(cmatrix_)
metrics.precision_recall_fscore_support(testY1, predY_)
```
