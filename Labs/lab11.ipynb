```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
from sklearn import datasets

iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features_standardized = scaler.fit_transform(features)
```

```python
from sklearn.decomposition import KernelPCA

# Create support vector classifier
pca = KernelPCA(n_components=2, kernel="rbf", gamma=15)
# Train model
model = pca.fit_transform(features_standardized)
```

```python
print("Original Features:", features_standardized.shape[1])
print("Reduced Features:", model.shape[1])
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def solve(model, X, y, model_func):
  scaler = StandardScaler()
  features_standardized = scaler.fit_transform(X)

  pca09 = PCA(n_components=0.9)
  model09 = pca09.fit_transform(features_standardized)
  trainX, testX, trainY, testY = train_test_split(model09, y, random_state=2, test_size=0.2)
  lr=model_func(random_state=2)
  lr.fit(trainX, trainY)
  predY = lr.predict(testX)
  print(f"\n{model} with n_components = 0.9")
  print("Accuracy Score", accuracy_score(testY, predY))
  print("Precision Score", precision_score(testY, predY, average='micro'))
  print("Recall Score", recall_score(testY, predY, average='micro'))
  print("F1 Score", f1_score(testY, predY, average='micro'))

  pca095 = PCA(n_components=0.95)
  model095 = pca095.fit_transform(features_standardized)
  trainX, testX, trainY, testY = train_test_split(model095, y, random_state=2, test_size=0.2)
  lr=model_func(random_state=2)
  lr.fit(trainX, trainY)
  predY = lr.predict(testX)
  print(f"\n{model} with n_components = 0.95")
  print("Accuracy Score", accuracy_score(testY, predY))
  print("Precision Score", precision_score(testY, predY, average='micro'))
  print("Recall Score", recall_score(testY, predY, average='micro'))
  print("F1 Score", f1_score(testY, predY, average='micro'))

  pca099 = PCA(n_components=0.99)
  model099 = pca099.fit_transform(features_standardized)
  trainX, testX, trainY, testY = train_test_split(model099, y, random_state=2, test_size=0.2)
  lr=model_func(random_state=2)
  lr.fit(trainX, trainY)
  predY = lr.predict(testX)
  print(f"\n{model} with n_components = 0.99")
  print("Accuracy Score", accuracy_score(testY, predY))
  print("Precision Score", precision_score(testY, predY, average='micro'))
  print("Recall Score", recall_score(testY, predY, average='micro'))
  print("F1 Score", f1_score(testY, predY, average='micro'))

  pca10 = PCA(n_components=10)
  model10 = pca10.fit_transform(features_standardized)
  trainX, testX, trainY, testY = train_test_split(model10, y, random_state=2, test_size=0.2)
  lr=model_func(random_state=2)
  lr.fit(trainX, trainY)
  predY = lr.predict(testX)
  print(f"\n{model} with n_components = 10")
  print("Accuracy Score", accuracy_score(testY, predY))
  print("Precision Score", precision_score(testY, predY, average='micro'))
  print("Recall Score", recall_score(testY, predY, average='micro'))
  print("F1 Score", f1_score(testY, predY, average='micro'))

  if len(X.columns) > 30:

    pca30 = PCA(n_components=30)
    model30 = pca30.fit_transform(features_standardized)
    trainX, testX, trainY, testY = train_test_split(model30, y, random_state=2, test_size=0.2)
    lr=model_func(random_state=2)
    lr.fit(trainX, trainY)
    predY = lr.predict(testX)
    print(f"\n{model} with n_components = 30")
    print("Accuracy Score", accuracy_score(testY, predY))
    print("Precision Score", precision_score(testY, predY, average='micro'))
    print("Recall Score", recall_score(testY, predY, average='micro'))
    print("F1 Score", f1_score(testY, predY, average='micro'))

    pca40 = PCA(n_components=40)
    model40 = pca40.fit_transform(features_standardized)
    trainX, testX, trainY, testY = train_test_split(model40, y, random_state=2, test_size=0.2)
    lr=model_func(random_state=2)
    lr.fit(trainX, trainY)
    predY = lr.predict(testX)
    print(f"\n{model} with n_components = 40")
    print("Accuracy Score", accuracy_score(testY, predY))
    print("Precision Score", precision_score(testY, predY, average='micro'))
    print("Recall Score", recall_score(testY, predY, average='micro'))
    print("F1 Score", f1_score(testY, predY, average='micro'))

    pca50 = PCA(n_components=50)
    model50 = pca50.fit_transform(features_standardized)
    trainX, testX, trainY, testY = train_test_split(model50, y, random_state=2, test_size=0.2)
    lr=model_func(random_state=2)
    lr.fit(trainX, trainY)
    predY = lr.predict(testX)
    print(f"\n{model} with n_components = 50")
    print("Accuracy Score", accuracy_score(testY, predY))
    print("Precision Score", precision_score(testY, predY, average='micro'))
    print("Recall Score", recall_score(testY, predY, average='micro'))
    print("F1 Score", f1_score(testY, predY, average='micro'))
```
