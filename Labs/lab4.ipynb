# Q1.

```python
import pandas as pd
data = pd.read_csv("Datasets/Heart_Dataset.csv")
data
```

```python
X = data.drop(columns='HeartDisease')
X
```

```python
Y = data['HeartDisease']
```

```python
X.info()
```

```python
X['Gender'].unique()
X['Gender'] = X['Gender'].replace('M', 1)
X['Gender'] = X['Gender'].replace('F', 0)
```

```python
X['ChestPainType'].unique()
X['ChestPainType'] = X['ChestPainType'].replace('ATA', 1)
X['ChestPainType'] = X['ChestPainType'].replace('NAP', 2)
X['ChestPainType'] = X['ChestPainType'].replace('ASY', 3)
X['ChestPainType'] = X['ChestPainType'].replace('TA', 4)
```

```python
X['RestingECG'].unique()
X['RestingECG']=X['RestingECG'].replace('Normal',1)
X['RestingECG']=X['RestingECG'].replace('ST',2)
X['RestingECG']=X['RestingECG'].replace('LVH',3)
```

```python
X['ExerciseAngina'].unique()
X['ExerciseAngina']=X['ExerciseAngina'].replace('N',0)
X['ExerciseAngina']=X['ExerciseAngina'].replace('Y',1)
```

```python
X['ST_Slope'].unique()
X['ST_Slope'] = X['ST_Slope'].replace('Up', 1)
X['ST_Slope'] = X['ST_Slope'].replace('Flat', 2)
X['ST_Slope'] = X['ST_Slope'].replace('Down', 3)
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0, test_size=0.2)
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape, sep='\n')
```

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
classifier1 = GaussianNB()
classifier1.fit(X_train, Y_train)
```

# Q2.

```python
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, roc_auc_score

pred_Y = classifier1.predict(X_test)
print("The accuracy is", accuracy_score(Y_test, pred_Y)*100, "%")
print("The recall score is", recall_score(Y_test, pred_Y)*100, "%")
print("The precision is", precision_score(Y_test, pred_Y)*100, "%")
print("The roc score is", roc_auc_score(Y_test, pred_Y)*100, "%")
confusion_matrix(Y_test, pred_Y)
```
