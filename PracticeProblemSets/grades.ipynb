## 2. Check for missing values and suggest methods to fix them.<br/>

We can replace the null values with grade 'F'. 

```python
X.info()
```

## 3. Which encoding technique would be most suitable to encode grades of the courses? Use it to convert the categorical columns into numeric ones.

We'll go with ordinal encoding here.

```python
mapp = {'A+': 1,
        'A': 2,
        'A-': 3,
        'B+': 4,
        'B': 5,
        'B-': 6,
        'C+': 7,
        'C': 8,
        'C-': 9,
        'D+': 10,
        'D': 11,
        'F': 12,
        'I': 13,
        'W': 14,
        'WU': 15}

def helper(grade):
    return mapp[grade]
    
```

```python

for column in X.columns:
    X[column] = X[column].apply(helper)
    
```
