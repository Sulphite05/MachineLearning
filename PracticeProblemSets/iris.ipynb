# Assignment 1: Visualising the Data

Download the Iris Dataset from sklearn toy datasets, the way Boston Dataset was downloaded in notebooks N3 and N4 (https://scikit-learn.org/stable/datasets/toy_dataset.html).

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
from sklearn.datasets import load_iris
```

## 1. Create a dataframe for this dataset.

```python
iris = load_iris()
iris_dict = dict()

for i, key in enumerate(iris.feature_names):
    iris_dict[key] = iris.data[:,i]
iris_dict['target'] = iris.target

data = pd.DataFrame(iris_dict)
data
```

## 2. Plot histograms for all the features and target.

```python
data.hist(figsize=(10,10))
```

## 3. Plot heatmap and identify correlations. Analyze the correlations for signs of multicollinearity.

```python
sns.heatmap(data.corr())
```

```python
data['petal width (cm)'].corr(data['petal length (cm)'])
```

Hence petal width and petal length are highly correlated showing signs of multicollinearity

## 4. Generate pairplot for the dataset.

```python
sns.pairplot(data)
```

## 5. Normalize the dataset and store the new normalized dataset in file Normalized_Iris_Dataset.csv.

```python
norm_data = data.copy()
norm_data['petal length (cm)'] = (data['petal length (cm)'] - data['petal length (cm)'].min()) / (data['petal length (cm)'].max() - data['petal length (cm)'].min())
norm_data['petal width (cm)'] = (data['petal width (cm)'] - data['petal width (cm)'].min()) / (data['petal width (cm)'].max() - data['petal width (cm)'].min())
norm_data['sepal length (cm)'] = (data['sepal length (cm)'] - data['sepal length (cm)'].min()) / (data['sepal length (cm)'].max() - data['sepal length (cm)'].min())
norm_data['sepal width (cm)'] = (data['sepal width (cm)'] - data['sepal width (cm)'].min()) / (data['sepal width (cm)'].max() - data['sepal width (cm)'].min())
norm_data
```

```python
norm_data.to_csv("Datasets/Normalized_Iris_Dataset.csv")
```
